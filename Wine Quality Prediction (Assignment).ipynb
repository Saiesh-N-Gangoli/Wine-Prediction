```python
#data preprocessing(seprate data and label)
A = wine_data.drop('quality', axis=1)
```

```python
print(A)
```

```python
#Label Binarization
Y = wine_data['quality'].apply(lambda y_value:1 if y_value>=7 else 0)
```

```python
print(Y)
```

```python
#spliting train and test data
A_train, A_test, Y_train, Y_test = train_test_split(A,Y, test_size=0.2,random_state=2)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

```python
#Training the model using random forest algorithm
model = RandomForestClassifier()
```

```python
model.fit(A_train, Y_train)
```

```python
#model evaluvation using accuracy score
#accuracy of test data
A_test_prediction = model.predict(A_test)
test_data_accuracy = accuracy_score(A_test_prediction, Y_test)
```

```python
td = test_data_accuracy*100
```

```python
print('Accuracy:', td)
```
